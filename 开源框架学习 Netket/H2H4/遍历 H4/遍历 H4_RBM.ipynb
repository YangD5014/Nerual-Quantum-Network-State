```python
import numpy as np
import matplotlib.pyplot as plt

# 加载之前保存的H4分子能量数据
data = np.load('/Users/yangjianfei/mac_vscode/神经网络量子态/VMC遍历H4/h4_energy_results_RBM_alpha1.npz')

# 提取键长和能量数据
bond_distances = np.linspace(0.5, 2.5, 25)
E_fci = data['E_fci']
E_vmc = data['E_vmc']

# 绘制H4分子的势能面(PES)
plt.figure(figsize=(10, 6))
plt.plot(bond_distances, E_fci, 'o-', label='FCI能量', color='blue')
plt.plot(bond_distances, E_vmc, 's-', label='VMC能量 (RBM)', color='red')
plt.xlabel('键长 (Å)')
plt.ylabel('能量 (Hartree)')
plt.title('H4分子势能面(PES)')
plt.grid(True)
plt.legend()
plt.show()

# 打印一些统计信息
print(f'键长范围: {bond_distances.min():.3f} - {bond_distances.max():.3f} Å')
print(f'FCI能量范围: {E_fci.min():.6f} - {E_fci.max():.6f} Hartree')
print(f'VMC能量范围: {E_vmc.min():.6f} - {E_vmc.max():.6f} Hartree')
print(f'平均绝对误差: {np.mean(np.abs(E_fci - E_vmc)):.6f} Hartree')
```
