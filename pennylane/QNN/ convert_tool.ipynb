```python
import re

def convert_latex_to_notion(text):
    """
    将文本中的 LaTeX 公式转换为 Notion 兼容格式
    处理以下情况：
    1. 行内公式 \(...\) → $...$
    2. 行间公式 \[...\] → $$...$$
    3. 环境公式 \begin{xxx}...\end{xxx} → $$...$$
    4. 确保已经正确转义的公式保持不变
    """
    # 处理行内公式
    text = re.sub(r'\\\((.*?)\\\)', r'$\1$', text)
    # 处理行间公式
    text = re.sub(r'\\\[(.*?)\\\]', r'$$\1$$', text)
    # 处理环境公式
    text = re.sub(r'\\begin\{.*?\}(.*?)\\end\{.*?\}', r'$$\1$$', text, flags=re.DOTALL)
    # 处理已经用 $ 包裹的公式，确保不重复包裹
    text = re.sub(r'\$(.*?)\$', lambda m: m.group(0) if m.group(1).startswith('$') else m.group(0), text)
    return text

# 示例文本
latex_text = """
VMC的目标是最小化能量期望值 \( \langle \hat{H} \rangle = \frac{\langle \Psi_\theta | \hat{H} | \Psi_\theta \rangle}{\langle \Psi_\theta | \Psi_\theta \rangle} \)，其中 \( |\Psi_\theta \rangle \) 是含参数 \( \theta \) 的试探波函数。梯度计算的核心是求解：
\[
\nabla_\theta \langle \hat{H} \rangle = \langle \nabla_\theta \ln \Psi_\theta \cdot (E_L - \langle \hat{H} \rangle) \rangle
\]
其中 \( E_L = \frac{\langle \Psi_\theta | \hat{H} | \Psi_\theta \rangle}{\langle \Psi_\theta | \Psi_\theta \rangle} \) 是局域能量。
"""

# 转换并输出
notion_text = convert_latex_to_notion(latex_text)
print(notion_text)
```
