```python
from plum import dispatch

@dispatch
def process(x: int):
    return f"处理整数: {x}"

@dispatch
def process(x: str):
    return f"处理字符串: {x.upper()}"

print(process(10))    # 输出: 处理整数: 10
print(process("hello")) # 输出: 处理字符串: HELLO
```

```python
import jax
import jax.numpy as jnp

# 1. 3D数组 [2,3,4] → 折叠前2轴(0,2) → [6,4]
x = jnp.ones((2, 3, 4))
y = jax.lax.collapse(x, 0, 2)
print(y.shape)  # (6, 4)

# 2. 折叠最后2轴：start=-2, stop=x.ndim → [2, 12]
y2 = jax.lax.collapse(x, -2, x.ndim)
print(y2.shape)  # (2, 12)

# 3. stop=None → 从start到最后全部折叠：[2,3,4] → [2, 12]
y3 = jax.lax.collapse(x, 1)  # start=1, stop=None
print(y3.shape)  # (2, 12)

# 4. 4D示例 [B, H, W, C] → 折叠H,W → [B, H*W, C]
x4 = jnp.ones((8, 32, 32, 3))
y4 = jax.lax.collapse(x4, 1, 3)
print(y4.shape)  # (8, 1024, 3)
```
